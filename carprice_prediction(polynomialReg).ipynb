```python
import numpy as np

x=df[['enginesize']]
y=df['price']
```

```python
df.dtypes
```

```python
#training the linear regression model on whole dataset
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
# Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

print(" Models Trained Successfully!")
```

```python
# Visualising the Linear Regression results
# Visualising the Linear Regression results
y_pred = model.predict(x)

plt.scatter(x,y, color='green')
plt.plot(x, model.predict(x), color = 'blue')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price (Linear Regression)')
plt.show()
```

```python
# Visualising the Polynomial Regression results
import numpy as np

# Sort X values
X_sorted = x.sort_values(by=x.columns[0])
X_poly_sorted = poly_reg.transform(X_sorted)

# Plot
plt.scatter(x, y, color='red')
plt.plot(X_sorted, lin_reg_2.predict(X_poly_sorted), color='blue')

plt.title('enginesize vs price (Polynomial Regression)')
plt.xlabel('enginesize')
plt.ylabel('price')
plt.show()


```

```python
from sklearn.metrics import r2_score,root_mean_squared_error,mean_absolute_error,mean_squared_error
y_poly_pred=lin_reg_2.predict(poly_reg.transform(x))
#polynomial
mse_poly=mean_squared_error(y,y_poly_pred)
mae_poly=mean_absolute_error(y,y_poly_pred)
r2_poly=r2_score(y,y_poly_pred)
rmse_poly=root_mean_squared_error(y,y_poly_pred)
print("MSE:",mse_poly)
print("MAE:",mae_poly)
print("R2:",r2_poly)
print("RMSE:",rmse_poly)
```
